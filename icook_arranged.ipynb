```python
# 取得目標網頁，將 response 的 html 用 bs4 解析轉成 BeautifulSoup 物件 
# 若不是mac，記得要在 Chrome 開發人員工具的 Network 中 Request Headers 找自己的 user-agent (見下一條)
# 丞優表示 user-agent 有寫就OK , 瀏覽器和作業系統平台不會影響
url = "https://icook.tw/categories"
headers = {"user-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36"
          }
resp = requests.get(url, headers=headers)
html = BeautifulSoup(resp.text)
```

```python
# 需跳過要vip會員的食譜
base_url = "https://icook.tw"
category = html.find_all("div", class_="inner-block")[2]


#清除不相關的標籤，回傳subcat = [("分類","下一頁的url")]
subcat = []
for sub in category.find_all("a"):
    try:
        pair = (sub["name"], base_url + sub["href"])
        subcat.append(pair)
    except:
        pass 
```

```python
# 把子分類 url 清單讀回來
import csv
subcat = []
with open("subcat_urls.csv", "r", encoding="utf-8", newline="") as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        subcat.append((row[0],row[1]))
```

```python
subcat
```

```python
recipe_url_list = []

# 蒐集所有次分類內的食譜清單的方法
def getRecipesByPage(html):
        for each_recipe in html.find_all("a", class_="browse-recipe-touch-link"):
            recipe_url = base_url + each_recipe["href"]
            recipe_url_list.append((subcat[subcat_index][0], recipe_url))

# 進入每個次分類頁面
# subcat共191個頁面，光食譜清單就要爬很久，測試時記得把清單縮小 e.g. subcat[:5]
subcat_index = 0
for each_subcat in subcat:
    subcat_entry_html = BeautifulSoup(requests.get(each_subcat[1], headers=headers).text)
    
    # 次分類中食譜的第一頁
    getRecipesByPage(subcat_entry_html)
     
    # 若有下一頁，翻頁後繼續取得
    html = subcat_entry_html
    ctrl = True
    while ctrl == True:
        try:
            if not html.find("a", rel="next")["href"] == None:
                next_pg_html = BeautifulSoup(requests.get(base_url + html.find("a", rel="next")["href"], headers=headers).text)
                getRecipesByPage(next_pg_html)
                html = next_pg_html
        except:
            ctrl = False
            print(subcat[subcat_index][0],"子分類爬取完畢")
        
    subcat_index += 1
```

```python
# 把 recipe_url_list 存起來
import csv
with open("recipe_url_list.csv", "w", encoding="utf-8", newline="") as csvfile:
    writer = csv.writer(csvfile)
    for each_recipe_url in recipe_url_list:
        writer.writerow([each_recipe_url[0],each_recipe_url[1]])
```

```python
# 把食譜存成有序dict : OrderedDict --> json用
# *** python 3.6 版本以上，dict 會記住元素的插入順序，大部分情況下不再需要使用OrderedDict ***
# from collections import OrderedDict

# 進入每個食譜
# 測試時記得把清單縮小 e.g. recipe_url_list[:5]
# ordered_recipe_list = []

# 準備食譜物件蒐集陣列
recipe_list = []

for each_recipe_url in recipe_url_list[:10]:
    recipe_html = BeautifulSoup(requests.get(each_recipe_url[1], headers=headers).text)
    
    # 測試行
    print(each_recipe_url)
    
    # 取得食譜名、食譜圖片網址、材料清單
    # recipe_name : str
    recipe_name = recipe_html.find("h1").text.strip()
    

    # img_src : str
    recipe_img_src = recipe_html.find("img", class_="main-pic")["src"]

    # ingredients : dict
    ingredients = recipe_html.find_all("div", itemprop="ingredients")
    ingredient_preparation = {}
    for ingre in ingredients:
        ingre_name = ingre.find("div", class_="ingredient-name").text
        ingre_unit = ingre.find("div", class_="ingredient-unit").text
        ingredient_preparation[ingre_name] = ingre_unit
    
    # recipe_catergory : str 
    recipe_category = each_recipe_url[0]
    
    #  做成一個 Recipe 物件{}
    a_recipe = {}
    a_recipe["食譜名稱"] = recipe_name
    a_recipe["料理圖片"] = recipe_img_src
    a_recipe["材料"] = ingredient_preparation
    a_recipe["食譜網址"] = each_recipe_url[1]
    a_recipe["食譜分類"] = recipe_category
    
#      #  做成一個 Ordered_Recipe 物件{}
#     an_ordered_recipe = OrderedDict() # 生成 OrderedDict 物件來儲存單份食譜
#     an_ordered_recipe["食譜名稱"] = recipe_name
#     an_ordered_recipe["料理圖片"] = recipe_img_src
#     an_ordered_recipe["材料"] = ingredient_preparation
#     an_ordered_recipe["食譜網址"] = each_recipe_url[1]
#     an_ordered_recipe["食譜分類"] = recipe_category
    
    # 存進食譜清單
    recipe_list.append(a_recipe)
#     ordered_recipe_list.append(an_ordered_recipe)
```
