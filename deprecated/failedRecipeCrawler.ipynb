```python
import sys
import requests
from bs4 import BeautifulSoup
from random import randint
from tools import crawlers
```

```python
with open("recipes_crawl_failure.json", "r", encoding="utf-8") as infile:
    recipe_str = infile.read()
    recipe_list = list(eval(recipe_str))
```

```python
crawled_recipe_list = []
for recipe in recipe_list:
    recipe_subcat_name = recipe[0]
    a_recipe = crawlers.getRecipeJson(recipe)
    if type(a_recipe) is dict:
        crawled_recipe_list.append(a_recipe)
    else:
        with open("recipes_crawl_failure_again.json", "a", encoding="utf-8") as failFile:
            a_recipe_str = repr(recipe)
            failFile.write(a_recipe_str + ",")    
```
